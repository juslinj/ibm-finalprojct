```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.metrics import jaccard_score
from sklearn.metrics import f1_score
from sklearn.metrics import log_loss
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score
import sklearn.metrics as metrics
```

```python
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillUp/labs/ML-FinalAssignment/Weather_Data.csv')

df.head()
```

```python
df_sydney_processed = pd.get_dummies(data=df, columns=['RainToday', 'WindGustDir', 'WindDir9am', 'WindDir3pm'])
```

```python
df_sydney_processed.replace(['No', 'Yes'], [0,1], inplace=True)
```

```python
df_sydney_processed.drop('Date',axis=1,inplace=True)
```

```python
df_sydney_processed = df_sydney_processed.astype(float)
```

```python
features = df_sydney_processed.drop(columns='RainTomorrow', axis=1)
Y = df_sydney_processed['RainTomorrow']
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_sydney_processed,Y,test_size=0.2,random_state=10)
```

```python
LinearReg = LinearRegression()
LinearReg.fit(x_train,y_train)
```

```python
predictions = LinearReg.predict(x_test)
```

```python
LinearRegression_MAE = metrics.mean_absolute_error(predictions,y_test)
LinearRegression_MSE = metrics.mean_squared_error(predictions,y_test)
LinearRegression_R2 = metrics.r2_score(predictions,y_test)
```

```python
Report = [{'Model': 'Linear Regression',
        'Accuracy': None,  # Not applicable for Linear Regression
        'Jaccard Index': None,  # Not applicable for Linear Regression
        'F1-Score': None,  # Not applicable for Linear Regression
        'LogLoss': None,  # Not applicable for Linear Regression
        'Mean Absolute Error': LinearRegression_MAE,
        'Mean Squared Error': LinearRegression_MSE,
        'R2 Score': LinearRegression_R2
    }
         ]
print(pd.DataFrame(Report))
```

```python
KNN = KNeighborsClassifier(n_neighbors=4)
KNN.fit(x_train,y_train)
```

```python
predictions = KNN.predict(x_test)
```

```python
KNN_Accuracy_Score = accuracy_score(predictions,y_test)
KNN_JaccardIndex = jaccard_score(predictions,y_test) 
KNN_F1_Score = f1_score(predictions,y_test)
```

```python
Tree = DecisionTreeClassifier()
Tree.fit(x_train,y_train)
```

```python
predictions = Tree.predict(x_test)
```

```python
Tree_Accuracy_Score = accuracy_score(predictions, y_test)
Tree_JaccardIndex = jaccard_score(predictions,y_test)
Tree_F1_Score = f1_score(predictions,y_test)
```

```python
x_train, x_test, y_train, y_test = train_test_split(features,Y,test_size=0.2,random_state=1)
```

```python
LR = LogisticRegression(solver='liblinear') 
LR.fit(x_train,y_train)
```

```python
predictions = LR.predict(x_test)
```

```python
LR_Accuracy_Score = accuracy_score(predictions,y_test)
LR_JaccardIndex = jaccard_score(predictions,y_test)
LR_F1_Score = f1_score(predictions,y_test)
LR_Log_Loss = log_loss(predictions,y_test)
```

```python
SVM = svm.SVC()
SVM.fit(x_train,y_train)
```

```python
predictions = SVM.predict(x_test)
```

```python
SVM_Accuracy_Score = accuracy_score(predictions,y_test)
SVM_JaccardIndex = jaccard_score(predictions,y_test)
SVM_F1_Score = f1_score(predictions,y_test)
```

```python
Report = [
    {
        'Model': 'SVM',
        'Accuracy': SVM_Accuracy_Score,
        'Jaccard Index': SVM_JaccardIndex,
        'F1-Score': SVM_F1_Score,
        'LogLoss': None  # Not applicable for SVM
    },
    {
        'Model': 'LR',
        'Accuracy': LR_Accuracy_Score,
        'Jaccard Index': LR_JaccardIndex,
        'F1-Score': LR_F1_Score,
        'LogLoss': LR_Log_Loss
    },
    {
        'Model': 'Decision Tree',
        'Accuracy': Tree_Accuracy_Score,
        'Jaccard Index': Tree_JaccardIndex,
        'F1-Score': Tree_F1_Score,
        'LogLoss': None  # Not applicable for Decision Tree
    },
    {
        'Model': 'KNN',
        'Accuracy': KNN_Accuracy_Score,
        'Jaccard Index': KNN_JaccardIndex,
        'F1-Score': KNN_F1_Score,
        'LogLoss': None  # Not applicable for KNN
    },
    {
        'Model': 'Linear Regression',
        'Accuracy': None,  # Not applicable for Linear Regression
        'Jaccard Index': None,  # Not applicable for Linear Regression
        'F1-Score': None,  # Not applicable for Linear Regression
        'LogLoss': None,  # Not applicable for Linear Regression
        'Mean Absolute Error': LinearRegression_MAE,
        'Mean Squared Error': LinearRegression_MSE,
        'R2 Score': LinearRegression_R2
    }
]

df = pd.DataFrame(Report)
print(df)
```
